```python
import pandas as pd


conditions = [
    ("maas", "Normal", ["dusuk", "yuksek"]),
    ("maas", "yuksek", ["dusuk", "Normal"]),
    ("maas", "dusuk", ["Normal", "yuksek"]),
    ("deneyim", "no", ["orta", "iyi"]),
    ("deneyim", "orta", ["yok", "iyi"]),
    ("deneyim", "iyi", ["yok", "orta"]),
    ("gorev", "uzman", ["yoneti"]),
    ("gorev", "yonetici", ["uzman"])
]

left_counts = [0] * 8
right_counts = [0] * 8

for i in range(11):
    condition = conditions[i]
    if i < 3:
        left_counts[i] += 1
        for right_condition in condition[2]:
            right_counts[right_condition.index(right_condition)] += 1
    elif i < 6:
        left_counts[i] += 1
        for right_condition in condition[2]:
            right_counts[right_condition.index(right_condition)] += 1
    elif i < 8:
        left_counts[i] += 1
        for right_condition in condition[2]:
            right_counts[right_condition.index(right_condition)] += 1

data = {
    "Left": left_counts,
    "Right": right_counts
}
df = pd.DataFrame(data)

print(df)
```

```python
import numpy as np

splits_matrix = {
    'maas=Normal': ['B_sol', 'B_sag'],
    'maas=yukse': ['B_sol', 'B_sag'],
    'maas=dusuk': ['B_sol', 'B_sag'],
    'deneyim=yok': ['B_sol', 'B_sag'],
    'deneyim=orta': ['B_sol', 'B_sag'],
    'deneyim=iyi': ['B_sol', 'B_sag'],
    'gorev=uzman': ['B_sol', 'B_sag'],
    'gorev=yonetici': ['B_sol', 'B_sag']
}

class_counts = {
    'Tsinif1': [10, 6, 4],
    'Tsinif2': [0, 0, 0]
}


best_split = None
best_suitability = float('-inf')


for key, value in splits_matrix.items():
    B_sol = np.sum(class_counts[value[0]])
    B_sag = np.sum(class_counts[value[1]])

    terim1 = 2 * (B_sol / np.sum(list(class_counts.values()))) * (B_sag / np.sum(list(class_counts.values())))
    terim2 = 0

    for j in range(3):
        terim2 += np.abs(class_counts['Tsinif'+str(j+1)][0] / B_sol - class_counts['Tsinif'+str(j+1)][1] / B_sag)

    suitability = terim1 * terim2

    if suitability > best_suitability:
        best_suitability = suitability
        best_split = key

print(f"En yüksek uygunluk değerine sahip bölünme: {best_split}")
```
