```python
import pandas as pd
from math import sqrt

oklid = []
siralama = []
x=7
y=3

data = {
    'X': [1, 2, 2, 3, 4, 5, 6, 8, 10, 11],
    'Y': [3, 5, 3, 9, 7, 2, 8, 6, 6, 1],
    'Z': ['Negatif', 'Pozitif', 'Pozitif', 'Negatif', 'Pozitif', 'Pozitif', 'Pozitif', 'Negatif', 'Negatif', 'Negatif']
}

df = pd.DataFrame(data)

for i in range(10):
    deger = sqrt((x-df.iloc[i,0])**2+(y-df.iloc[i,1])**2)
    oklid.append(deger)
    
data2 = {
    'X': [1, 2, 2, 3, 4, 5, 6, 8, 10, 11],
    'Y': [3, 5, 3, 9, 7, 2, 8, 6, 6, 1],
    'mesafe': oklid
} 

df2 = pd.DataFrame(data2)

for j in range(10):
    z=1
    for x in range(10):
        if df2.iloc[j,2]>df2.iloc[x,2]:
            z+=1
    siralama.append(z)
    

df2['siralama'] = siralama    
df2
```
